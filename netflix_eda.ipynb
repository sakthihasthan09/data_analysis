```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
```

```python
netflix_data = pd.read_csv("./warmup-data/netflix_titles.csv")
```

```python
netflix_data.head()
```

```python
netflix_data.fillna('Unknown', inplace=True)
print(netflix_data.isnull().sum())
```

```python
netflix_data_copy=netflix_data.copy()
```

```python
#add new columns
netflix_data_copy.head()
```

```python
print(netflix_data_copy.isnull().sum())
```

```python
netflix_data_copy['duration_int']= netflix_data_copy["duration"].str.extract(r"(\d+)").astype(float)
netflix_data_copy['duration_type']= netflix_data_copy["duration"].str.extract(r"([a-zA-Z]+)").astype(str).apply(lambda x: x.str.strip())
netflix_data_copy.head()
```

```python
print(netflix_data_copy.columns)
```

```python
# Example: from a date_added column
netflix_data_copy['year'] = pd.to_datetime(netflix_data_copy['date_added'], errors='coerce').dt.year
netflix_data_copy['year'] = netflix_data_copy['year'].fillna(0).astype(int)

```
